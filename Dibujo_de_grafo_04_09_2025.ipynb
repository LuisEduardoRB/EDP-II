#Dibujo de redes
Clase 04/09/2025
Grafo de 4 nodos y 5 Arcos.
En el siguiente programa vemos como es que podemos imprimir la imagen de una red, asignarle coordenadas, añadir nodos, añadir arcos e imprimir cada uno de ellos

```python
import networkx as nx
#Creamos el grafo dirigido vacío
G=nx.DiGraph()
#Creamos el vector que contiene todos los arcos y nodos
arcos=[("A","B"), ("A","C"), ("B","C"), ("B","D"), ("C","D")]
#Este tipo de formato ya da un sentido al arco
#Añadimos los arcos y nodos al grafo vacío
G.add_edges_from(arcos)
#le asignamos coordenadas fijas a los nodos
pos= {"A":(-1,0), "B":(0,-1), "C":(0,1), "D":(1,0)}
#Mandamos a imprimir la imagen del grafo junto con la posición asignada previamente
nx.draw_networkx(G,pos)
#Con las siguientes funciones mandamos a imprimir cada uno de los elementos de la red
#Primero se pone la red de la que la estamos sacando, un punto y lo que queremos imprimir
print("Nodos: ",G.nodes())
print("Arcos: ", G.edges())
#Con la siguiente función vemos el grado de cada nodo:
print("Grado nodo : ",G.degree("A"))
print("Grado nodo : ",G.degree("B"))
print("Grado nodo : ",G.degree("C"))
print("Grado nodo : ",G.degree("D"))
```
